```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:/Users/Lenovo/OneDrive/Desktop/congnifiz/Dataset .csv')
print(df.columns)
```

```python
restaurant_chain = df.groupby("Restaurant Name").size().reset_index(name="OutLetCount")
new = restaurant_chain[restaurant_chain["OutLetCount"] > 1]
restaurant_chains = new.sort_values(by="OutLetCount", ascending=False)
print(restaurant_chains[["Restaurant Name", "OutLetCount"]].head(10))
```

```python
plt.bar(restaurant_chains["Restaurant Name"][:5],restaurant_chains["OutLetCount"][:5])
plt.title("Top 5 Restarunt Chain By Number Of Outlets")
plt.xlabel("Restaurant_chain")
plt.ylabel("Number Of Outlets")
plt.show()
```

```python
ratings=df.groupby("Restaurant Name")["Aggregate rating"].mean().reset_index(name="Aggregate rating").sort_values(by="Aggregate rating",ascending=False)
print(ratings)
```

```python
votes= df.groupby("Restaurant Name")["Votes"].mean().reset_index(name="Votes").sort_values(by="Votes",ascending=False)
print(votes)
```

```python
plt.bar(ratings["Restaurant Name"][:5],ratings["Aggregate rating"][:5])
plt.title("Top 5 Restarunt Chain By Average Rating")
plt.xlabel("Restaurant_chain")
plt.ylabel("Average Rating")
plt.show()
```

```python
plt.bar(votes["Restaurant Name"][:5],votes["Votes"][:5])
plt.title("Top 5 Restarunt Chain By Total Votes")
plt.xlabel("Restaurant_chain")
plt.ylabel("Average Rating")
plt.show()
```
