```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/Lenovo/OneDrive/Desktop/congnifiz/Dataset .csv")
df.head()
```

```python
print(df.columns)
```

```python
Aggregate_rating_distribution=df['Aggregate rating'].value_counts()
print(Aggregate_rating_distribution)
```

```python
a=Aggregate_rating_distribution.idxmax()
print('Most Common Rates',a)
```

```python
plt.hist(df['Aggregate rating'],color='yellow',edgecolor='black')
plt.title('distribution of aggregate ratings')
plt.xlabel('Aggregate rating')
plt.ylabel('Counts')
plt.show()
```

```python
voting=df['Votes'].value_counts()
print(voting)
```

```python
avg_voting=round(df['Votes'].mean(),2)
print('Average Number of Votes Received by Restaurants : ',avg_voting)
```
