```python

import matplotlib.pyplot as plt
import pandas as pd
```

```python
# DETERMINE THE TOP 3 MOST 
# CUISINES IN THS DATA SET
```

```python
df=pd.read_csv(r"C:/Users/Lenovo/OneDrive/Desktop/congnifiz/Dataset .csv")
df.head() 
```

```python
print(df.shape)
print(df.columns)
```

```python
print(df['Cuisines'])
```

```python
cuisine_count=df['Cuisines'].str.split(', ').explode('Cuisines').value_counts()
print(cuisine_count)
top_cuisines=cuisine_count.head(3)
print("TOP 3 Cuisines:",top_cuisines)
```

```python
colors=['green','red','yellow']
plt.bar(top_cuisines.index,top_cuisines.values,color=colors)
plt.title('TOP 3 CUISINES')
plt.xlabel('CUISINES')
plt.ylabel('COUNT')
plt.show()
```

```python
# Calculate the percentage of
# restaurants that serve each of the top
# cuisines.
```

```python
total_restaurant=len(df)
print(total_restaurant)
total_restaurants1=cuisine_count.head()
percentage=(total_restaurants1/total_restaurant)*100
print('THE MARKET SHARE OF THE TOP 5 CUISINES : ',percentage)
```

```python
plt.bar(total_restaurants1.index,percentage.values,color=colors)
plt.title('TOP 5 CUISINES %')
plt.xlabel('CUISINES')
plt.ylabel('PERCENTAGE')
plt.show()
```
