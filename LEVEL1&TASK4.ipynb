```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/Lenovo/OneDrive/Desktop/congnifiz/Dataset .csv")
df.head()
```

```python
print(df.columns)
```

```python
online_delivery=df['Has Online delivery'].value_counts().get('Yes',0)
print('Total Restaurants in online delivery : ',online_delivery)
```

```python
no=df['Has Online delivery'].value_counts().get('No',0)
print('Total Resstaurants in No online delivery : ',no)
```

```python
# Determine the percentage of restaurants that offer online delivery
row_count=len(df)
percentage_online_delivery=round((online_delivery/row_count)*100,2)
print('The pecentage of restaurant online delivery : ',percentage_online_delivery)
```

```python
avg_rating_yes=round(df[df['Has Online delivery']=='Yes']['Aggregate rating'].mean(),2)
print('Average Ratng with Online Delivery : ',avg_rating_yes)
```

```python
avg_rating_no=round(df[df['Has Online delivery']=='No']['Aggregate rating'].mean(),2)
print('Average Ratng without Online Delivery : ',avg_rating_no)
```

```python
labels=['Yes','No']
average_rating=[avg_rating_yes,avg_rating_no]
colors='blue','green'
plt.bar(labels,average_rating,color=colors)
plt.xlabel('ONLINE RATING')
plt.ylabel('AVERAGE RATING')
plt.title("Comparsion of Average Online Deliery ")
plt.show()
```
